### apply function in dataframe in entire column

```python
df_sales = pd.read_csv(r'C:\Users\ss\Downloads\sales_data_final.csv')
```

```python
df_sales.head()
```

```python
def profit_flag(a):
    if a >= 0:
        return 'positive'
    else:
        return 'negative'
```

```python
profit_flag(-3)
```

```python
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag)
```

```python
df_sales.head()
```

### create new column to store length of customer name

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

### create a flag divide quantity into high, low, medium

```python
def quantity_flag(a):
    if a < 10:
        return 'low'
    elif a >10 and a< 20 :
        return 'medium'
    else:
        return 'high'
```

```python
df_sales['flag_quantity'] = df_sales['quantity'].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
```
